```python
# Import Libraries

import pandas as pd
```

```python
# Global function definitions

def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0
```

```python
# Collect/Read dataset

df = pd.read_csv('kl.csv', encoding='windows-1252')
# df.shape
# df.describe()
# df.values
df
```

```python
# Cleaning the data: Find the MVP Player

df1 = pd.DataFrame(df, columns = ['Name','Value', 'Wage','Age'])
wage = df1['Wage'].replace('[\€,]','',regex=True).apply(value_to_float)
value = df1['Value'].replace('[\€,]','',regex=True).apply(value_to_float)
df1['difference'] = value - wage
df1['Wage_num'] = wage
df1['Value_num'] = value
df1.sort_values('difference', ascending=False)
```
